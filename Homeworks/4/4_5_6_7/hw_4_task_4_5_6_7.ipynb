```python
import numpy
import matplotlib.pyplot
import math
```

```python
target_plot_points=100
training_points=2
tests=1000
x_target_points=numpy.linspace(-1,1,target_plot_points)
y_target_points=[target_function(i) for i in x_target_points]

def target_function(x):
    return math.sin(math.pi * x)

def plot_target_function(target_plot_points):
    matplotlib.pyplot.plot(x_target_points,y_target_points,label='sin(pi*x)')

def show_plot():
    matplotlib.pyplot.legend()
    matplotlib.pyplot.axis([-1.001,1.001,-2,2])
    matplotlib.pyplot.show()
```

```python
def linear_regression_ax(training_points):
    X=numpy.random.uniform(-1,1,training_points)
    Y=[target_function(i) for i in X]
    X=numpy.reshape(X,(training_points,1))
    X=numpy.hstack((numpy.zeros((training_points,1)),X))
    Y=numpy.reshape(Y,(training_points,1))
    X_inv = numpy.linalg.pinv(X)
    w=numpy.dot(X_inv,Y)
    return w

g_average=0.
for times in range(tests):
    w=linear_regression_ax(training_points)
    matplotlib.pyplot.plot(x_target_points,[i*w[1] for i in x_target_points],color='yellow')
    g_average+=w[1]/tests
print('g_average= ',g_average)

bias=0.
for x_point in x_target_points:
    target_y=target_function(x_point)
    g_average_y=g_average*x_point
    bias+=math.pow(g_average_y-target_y,2)/target_plot_points
print('bias= ',bias)

variance=0.
for x_point in x_target_points:
    w=linear_regression_ax(training_points)
    g_y=w[1]*x_point
    g_average_y=g_average*x_point
    variance+=math.pow(g_y-g_average_y,2)/target_plot_points
print('variance',variance)

print('E_out=variance+bias: ',variance+bias)

plot_target_function(target_plot_points)
show_plot()
```

```python
def linear_regression_b(training_points):
    X=numpy.random.uniform(-1,1,training_points)
    Y=[target_function(i) for i in X]
    X=numpy.hstack((numpy.ones((training_points,1)),numpy.zeros((training_points,1))))
    Y=numpy.reshape(Y,(training_points,1))
    X_inv = numpy.linalg.pinv(X)
    w=numpy.dot(X_inv,Y)
    return w

g_average=0.
for times in range(tests):
    w=linear_regression_b(training_points)
    matplotlib.pyplot.plot(x_target_points,[w[0] for i in x_target_points],color='yellow')
    g_average+=w[0]/tests
print('g_average= ',g_average)

bias=0.
for x_point in x_target_points:
    target_y=target_function(x_point)
    g_average_y=g_average
    bias+=math.pow(g_average_y-target_y,2)/target_plot_points
print('bias= ',bias)

variance=0.
for x_point in x_target_points:
    w=linear_regression_b(training_points)
    g_y=w[0]
    g_average_y=g_average
    variance+=math.pow(g_y-g_average_y,2)/target_plot_points
print('variance',variance)

print('E_out=variance+bias: ',variance+bias)

plot_target_function(target_plot_points)
show_plot()
```
