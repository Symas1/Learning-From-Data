```python
import numpy
import matplotlib.pyplot
import math
```

```python
def target_function(x):
    return math.sin(math.pi * x)

def plot_target_function(target_plot_points):
    

def linear_regression(training_points):
    X=[]
    Y=[]
    for i in range(training_points):
        x_point = numpy.random.uniform(-1,1,1).tolist()
        y_point = target_function(x_point[0])
        Y.append(y_point)
        X.append(numpy.array([x_point[0]],ndmin=2))
    X=numpy.reshape(X,(training_points,1))
    Y=numpy.reshape(Y,(training_points,1))
    X_inv = numpy.linalg.pinv(X)
    w = numpy.dot(X_inv,Y)
    return w[0]
```

```python
target_plot_points=1000
x = numpy.linspace(-1,1,target_plot_points)
y = [target_function(i) for i in x]

training_points=2

tests=100
g_avg=0
for i in range(tests):
    w=linear_regression(training_points)
    
    test_y = [i*w for i in x]
    matplotlib.pyplot.plot(x,test_y,color='yellow')

    g_avg+=w

g_avg/=tests
matplotlib.pyplot.plot(x,[i*g_avg for i in x],color='red',label='g_average y=ax')
print('g_avg= ',g_avg)

bias = 0
for i in range(target_plot_points):
    target_y = target_function(x[i])
    bias_y = g_avg*x[i]
    bias+=math.pow(bias_y-target_y,2)
bias/=target_plot_points
print('bias= ',bias)

variance=0
for i in range(target_plot_points):
    w=linear_regression(training_points)
    variance+=math.pow((w*x[i])-(g_avg*x[i]),2)
variance/=target_plot_points
print('variance= ',variance)

matplotlib.pyplot.plot(x,y,label='sin(pi*x)')
# for i in range(training_points):
#     matplotlib.pyplot.plot(X[i][0],Y[i][0],'.r')

matplotlib.pyplot.legend()
matplotlib.pyplot.axis([-1.001,1.001,-2,2])
matplotlib.pyplot.show()
```
