```python
import numpy
import matplotlib.pyplot
%matplotlib inline
```

```python
coins=1000
tests=10
total_tests=100000
list_v1 = []
list_v2 = []
list_v3 = []
for total in range(total_tests):
    # 1 - head, 0 - tail
    results=[]
    for i in range(coins):
        flips = numpy.random.randint(0,2,tests)
        unique, counts = numpy.unique(flips,return_counts=True)
        d = dict(zip(unique,counts))
        results.append(d)

    min_index=0
    for i in range(1,len(results)):
          if results[i].get(1,0)/(results[i].get(0,0)+results[i].get(1,0)) < results[min_index].get(1,0)/(results[min_index].get(0,0)+results[min_index].get(1,0)):
                        min_index=i
    c1 = results[0]
    c2 = results[numpy.random.randint(0,len(results))]
    c3 = results[min_index]
    v1 = c1.get(1,0)/(c1.get(1,0)+c1.get(0,0))
    v2 = c2.get(1,0)/(c2.get(1,0)+c2.get(0,0))
    v3 = c3.get(1,0)/(c3.get(1,0)+c3.get(0,0))
    list_v1.append(v1)
    list_v2.append(v2)
    list_v3.append(v3)
list_v1=numpy.around(list_v1,decimals=1)
list_v2=numpy.around(list_v2,decimals=1)
list_v3=numpy.around(list_v3,decimals=1)
hist_bins=numpy.linspace(0,1,30)
matplotlib.pyplot.hist(list_v1,bins=hist_bins)
matplotlib.pyplot.show()
matplotlib.pyplot.hist(list_v2,bins=hist_bins)
matplotlib.pyplot.show()
matplotlib.pyplot.hist(list_v3,bins=hist_bins)
matplotlib.pyplot.show()
```

```python
max_eps=3
list_epsilon=numpy.linspace(0.,max_eps,1000).tolist()
list_probability=[]
list_formula=[2*numpy.exp(-2*eps**2*tests) for eps in list_epsilon]
```

```python
list_probability.clear()
for eps in list_epsilon:
    count_less=0
    count_more=0
    for i in list_v1:
        if numpy.fabs(i-0.5)>eps:
            count_more+=1
        else:
            count_less+=1
    list_probability.append(count_more/(count_less+count_more))
matplotlib.pyplot.plot(list_epsilon,list_probability,'g.')
matplotlib.pyplot.plot(list_epsilon,list_formula)
matplotlib.pyplot.axis([0,max_eps,0,1])
matplotlib.pyplot.ylabel('P[v-u]>eps')
matplotlib.pyplot.xlabel('epsilon')
matplotlib.pyplot.show()
```

```python
the main point of the exercise is to realize that if you pick a coin carefully based on the data of the flips (in this case "carefully" means having minimum nu), then the distribution of the nu you see will not be what you would expect if you tossed that *same* coin again. If you tossed that c_min again, you expect to see a binomial distribution for heads. But the distribution of nu_min is clearly not binomial.

Continuing to learning, if you pick a hypothesis "carefully", say having minimum Ein, then the Ein you get will not necessarily reflect the distribution of errors you get when you "toss that hypothesis again" -- i.e test it on new data.
So v1 and v2 have binomial distributions and v3 has other distribution
```
